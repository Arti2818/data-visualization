```python
# Pearson Correlation between  all the Features
features_corr = my_data.corr()

plt_corr= plt.figure(figsize= (10, 8))
plt_corr = sns.heatmap(features_corr, linewidth= 0.1, cmap= "Blues")
plt_corr.set_title("The Pearson Correlation between Features",
                   fontsize= 15,
                   pad= 12)

plt.show()
```

```python
from sklearn.model_selection import train_test_split

x = my_data[my_data
    .columns.drop(['Fruits','Veggies','NoDocbcCost', 'HeartDiseaseorAttack'])].values
y = my_data['HeartDiseaseorAttack'].values



x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 1, stratify=y)
print('\n The total number of training dataset', x_train.shape)
print('\n The total of test dataset', x_test.shape)
```

```python
from sklearn.metrics import accuracy_score, r2_score
from sklearn.tree import DecisionTreeClassifier


my_model = DecisionTreeClassifier()
my_model.fit(x_train, y_train)

#now let's predict the model
y_predict_train = my_model.predict(x_train)

#now let's predict the model
y_pred = my_model.predict(x_test)


#compute train set accuracy
model_acc_train = accuracy_score(y_train, y_predict_train)
print("\nAccuracy on train set: {:.2f}\n".format(model_acc_train))
#compute test set accuracy
model_acc_test = accuracy_score(y_test, y_pred)
print("\nAccuracy on test set: {:.2f}\n".format(model_acc_test))
```

```python
import numpy as np
print("Mean absolute error (MAE): %.2f" %np.mean(np.absolute(y_pred - y_test)))
print("Mean squared error (MSE): %.2f" %np.mean((y_pred - y_test) ** 2))
print("R2-score (Coefficient of Determination):%.2f" % r2_score(y_pred , y_test) )
```

```python
from sklearn.ensemble import RandomForestClassifier
my_model_2 = RandomForestClassifier()
my_model_2.fit(x_train, y_train)
y_pred = my_model_2.predict(x_test)
print("Random Forest Classifier Accuracy: ",accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import roc_auc_score, roc_curve

tpr, fpr, th = roc_curve(y_test,my_model_2 .predict(x_test))
plt.plot(tpr,fpr)
plt.title(roc_auc_score(y_test, my_model_2.predict(x_test)))
plt.show()
```

```python
from sklearn.metrics import classification_report
print("Classification Report")
print(classification_report(y_test, y_pred))
```
